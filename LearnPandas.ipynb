```python
movies_url = 'http://bit.ly/imdbratings'

movies = pd.read_table(movies_url,sep=',')
movies.head()
```

```python
movies.title.sort_values(ascending=False)
```

```python
columns = ['content_rating', 'duration']

movies.sort_values(columns, ascending=True, inplace=True)

movies.head()
```

```python
movies.shape
```

```python
# create a list of movies that has same row with 
booleans = []
for length in movies.duration:
    if length > 200:
        booleans.append(True)
    else:
        booleans.append(False)

len(booleans)
```

```python
# same thing can be done with
is_long = movies.duration > 200
len(is_long)
```

```python
type(is_long)
```

```python
movies[is_long].sort_values('duration',ascending=False)
```

```python
# Additional --  we want to study the duration and only the genre of the movie 
movies[movies.duration >= 200].genre
# this is date frame we can use . and bracket 
```

```python
#best practive is to use loc function 
movies.loc[movies.duration >= 200].genre
```

```python
#multiple factor refactoring
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
#slow method to fetch data with OR
movies[(movies.genre == 'Crime') | (movies.genre == 'Drama') |( movies.genre == 'Action') ]
```

```python
#fast method of OR filtering
movies[movies.genre.isin(['Action', 'Drama','Crime'])]
```
