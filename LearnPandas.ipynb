```python
import pandas as pd
```

```python
# reads data in tabular format
url = "https://bit.ly/chiporders"

# read table reads tab separated values, the data needs to have a header column
orders = pd.read_table(url)
orders.head()
```

```python
#read data
url1 = 'https://bit.ly/uforeports'
ufo = pd.read_table(url1, sep=',')

ufo.head()
```

```python
#selecting a column
ufo['City']
# also works
ufo.City
# NOTE: column name is case sensitive "ufo.city" will give error
```

```python
#create a new column called location which concatenates city and state

ufo['Location'] = ufo['City'] + ', ' + ufo['State']
```
