Selecting multiple rows and columns from a pandas DataFrame
.loc
.iloc
.ix

```python
import pandas as pd
```

```python
url = 'http://bit.ly/uforeports'
ufos = pd.read_csv(url)
ufos.head(3)
```

.loc Usage
This is realy powerful and flexiable method

```python
ufos.loc[0,:]
```

```python
ufos.index
```

```python
ufos.loc[0:2,:]
```

```python
ufos.loc[0:2]
```

```python
# all rows
# columns :city
(ufos.loc[:,'City'])[0:5]
```

```python
ufos.loc[0:2,'City':'State']
```

```python
ufos[ufos.City == 'Oakland']
```

```python
# easier to read code
# here you specify the rows and columns you want
# ufo.loc[row, column]

ufos.loc[ufos.City == 'Oakland', :]
```

```python
# instead of showing all the column
ufos.loc[ufos.City =="Oakland", 'State']
```

```python
    ufos.loc[:,['City','State']]
```
